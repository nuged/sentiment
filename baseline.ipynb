```python
import pymorphy2
morph = pymorphy2.MorphAnalyzer()
```

```python
pos = []
neg = []

with open('rusentilex_2017.txt', 'r') as f, open('pos_nouns.txt', 'w') as g, open('neg_nouns.txt', 'w') as h:
    for line in f.readlines():
        if line.startswith('!') or line.strip() == '':
            continue
        line = line.strip().split(', ')
        if line[3] not in ['positive', 'negative']:
            continue
        if line[1] == 'Noun':
            flag = False
            for parse in morph.parse(line[0])[:1]:
                if 'anim' in parse.tag:
                    flag = True
                    break
            if not flag:
                continue
            if line[3] == 'positive':
                if line[0] not in pos:
                    pos.append(line[0])
            else:
                if line[0] not in neg:
                    neg.append(line[0])
```

```python
with open('pos_nouns.txt', 'w') as g, open('neg_nouns.txt', 'w') as h:
    for word in pos:
        g.write(word + '\n')
    for word in neg:
        h.write(word + '\n')
```
